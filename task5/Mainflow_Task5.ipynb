```python
questions=["1. How many people have heart disease problem and how many doesn't have heart problem? ",
           "2. People of which sex has most heart problem? ",
           "3. People of which sex has which types of chest pain most? ",
           "4. People with which chest pain are most prone to have heart failure?"]
questions
```

```python
# 1 How many people have heart disease problem and how many people doesn't have heart problem? "
df.target.value_counts()
```

In this dataset of 1025 records 526 people has a heart problem and 499 doesn't have heart problem.

```python
#plotting bar chart.
# 1 = Heart Disease, 0 = No Heart Disease
df.target.value_counts().plot(kind="bar", color=["red", "green"])
plt.title("Heart Disease vs No Heart Disease")
plt.xlabel('Presence of Heart disease')
plt.ylabel('Count')
```

```python
#plotting a pie chart
df.target.value_counts().plot(kind="pie", figsize=(6, 6))
plt.title("Heart Disease vs No Heart Disease")
plt.legend(["Heart disease", "No Heart disease"])
```

```python
# '0' in sex column represents 'Female'
# '1' in sex column represents 'Male'
# '0' in target column represents 'No Disease problem'
# '1' in target column represents 'Disease problem'
# Let's check how many 'Male' and 'female' are in the dataset

df.sex.value_counts()
```

```python
# Plotting a pie chart
df.sex.value_counts().plot(kind="pie", figsize=(6, 6))
plt.title("Male Female Ratio")
plt.legend(["Male", "Female"])
```

```python
#2. People of which sex has most heart problem?

# crosstab is used to analyse the relationship between two or more categorical varables
pd.crosstab(df.target, df.sex)
```

```python
sns.countplot(x="target", data = df, hue="sex")
plt.title("Heart problems variation according to sex")
plt.legend()
plt.xlabel("Male                  Female")
```

```python
# Number of male is  more than double in our dataset than female.
# More than '45% male' has heart problem and '75% female' has heart problem.
```

```python
#3. People of which sex has which types of chest pain most?
#counting values for different chest pain.

df.cp.value_counts()
```

```python
#plotting a bar chart
df.cp.value_counts().plot(kind="bar")
plt.title("Chest pain type vs count")
```

```python
pd.crosstab(df.cp, df.target)
```

```python
pd.crosstab(df.sex, df.cp).plot(kind="bar")
plt.title("Type of chest pain for Male vs Female")
plt.legend()
plt.xlabel("Female                       Male")
```

```python
#Most of 'Male' has 'type 0' chest pain and least of 'Male' has 'type 4' chest pain.
#In case of 'Female', 'type 0' and 'type 2' pain presence percentage is almost same.
```

```python
#4. People with which chest pain are most prone to have heart failure?
pd.crosstab(df.cp, df.target)
```

```python
sns.countplot(x='cp', data=df, hue='target')
```

```python
#Most of people who has 'type 0' chest pain has less chance of heart failure and we see the opposite for other types.

#create a distribution plot with normal distribution curve
#displot is an advance plot whish is in seaborn library,
#it combies a histogram, it display the distribution of the single numerical value,
#with curves represents'kde' = kernal density estimates,
#and providing like a comprensive view of the data distribution.
sns.displot(x= 'age', data = df, bins = 30, kde = True);
```

```python
#'58-59' year old people are most in dataset.
#let's plot another distribution plot for 'Maximum Heart Rate'
sns.displot(x= 'thalach', data = df, bins = 30, kde = True, color = 'brown');
```
