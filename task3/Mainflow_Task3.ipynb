```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading data


```python
df = pd.read_csv('householdtask3.csv')
```

```python
display(df.head(10))
```

```python
df.shape
```

# Scatter Plot

```python
plt.scatter(df['year'], df['own']) # year vs own column scatter plot
# Add title to scatter plot
plt.title('Scatter plot Year vs Own')
# Add x and y labels to plot
plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

# Line chart

```python

# Plot line chart
plt.figure(figsize=(10, 5))
plt.plot(df['year'], df['income'], marker='o', label='Income', color='r')
plt.plot(df['year'], df['expenditure'], marker='s', label='Expenditure', color='m')

plt.xlabel('Year')
plt.ylabel('Amount')

plt.title('Line Chart Income vs Expenditure')
plt.legend()

plt.show()
```

# Bar Chart

```python
# Bar Chart

plt.figure(figsize=(14, 7))

# Plotting total households and owner households with year as x-axis
plt.bar(df['year'] - 0.2, df['tot_hhs'], width=0.4, label='Total Households', color='b')
plt.bar(df['year'] + 0.2, df['own'], width=0.4, label='Owner Households', color='g')

# Adding x and y labels to the Bar chart
plt.xlabel('Year')
plt.ylabel('Number of Households')

# Add title to the chart
plt.title('Bar Chart Total vs Owner Households')

plt.legend() # Add legend

plt.show()
```

```python
import pandas as pd
# reading the database
data=pd.read_csv('tips.csv')
# printing the top 10 rows
display(data.head(10))
```

# Scatter Plot

```python
plt.scatter(x=data['day'], y=data["tip"],c=data['total_bill'])
# Adding Title to the Plot
plt.title ( "Scatter Plot-Day vs Tip" )
# Setting the X and Y labels
plt.xlabel(' Day')
plt.ylabel ('Tip')

plt.colorbar ()
plt.show()
```

# Line chart

```python
x=plt.plot (data['tip'])
y=plt.plot (data['size'])
# Adding Title to the Plot
plt.title( "Line chart-Tip vs Size")
# Setting the X and Y labels
plt.xlabel('Tip' )
plt.ylabel('Size' )
plt.show()
```

# Bar Chart

```python
plt.bar (data['day'], data ['tip'])
plt.title("Bar Chart- Day vs Tip")
# Setting the X and Y labels
plt.xlabel ('Day' )
plt.ylabel( 'Tip')
# Adding the legends
plt. show()
```
