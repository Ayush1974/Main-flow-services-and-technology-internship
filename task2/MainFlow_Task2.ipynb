```python
data1=data['Price']
data1
```

```python
import re
for j in range(len(data)):
    res = re.sub("\D", "", data1[j])
    print("The digits string is : ",float(str(res))/100)
    data1[j]=float(str(res))/100
```

```python
data['Price']=data1
data
```

```python
Q1=data['Price'].quantile(0.25)
Q3=data['Price'].quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
data=data.drop(['Book Name','Author Name'], axis=1)
```

```python
data=data[~((data<(Q1-1.5*IQR))|(data>(Q3+1.5*IQR))).any(axis=1)]
data
```
